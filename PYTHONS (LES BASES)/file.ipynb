```python
import pandas as pd

# Données d'exemple
df = pd.DataFrame({
    'mois': ['Jan', 'Feb', 'Mar', 'Apr'],
    'ventes': [100, 150, 130, 170]
})

# Graphique en ligne
df.plot(x='mois', y='ventes', kind='line', title="Ventes mensuelles")
```

```python
# Graphique en barres
df.plot(x='mois', y='ventes', kind='bar', color='skyblue', title="Ventes mensuelles")
```

```python
# Nuage de points
df.plot(x='mois', y='ventes', kind='scatter', title="Ventes par mois", color='purple')
```

```python
# Création d'un DataFrame
df = pd.DataFrame({
    'Catégorie': ['A', 'B', 'C', 'D'],
    'Valeurs': [50, 30, 15, 5]
})

# Diagramme circulaire pour la colonne 'Valeurs'
df.plot(kind='pie', y='Valeurs', labels=df['Catégorie'], autopct='%1.1f%%', title="Répartition des valeurs par catégorie")
```

```python
import numpy as np
# Création d'un DataFrame
df = pd.DataFrame({
    'Valeurs': np.random.normal(50, 10, 1000)
})

# Histogramme de la colonne 'Valeurs'
df.plot(kind='hist', y='Valeurs', bins=20, color='purple', alpha=0.6, title="Histogramme de la colonne 'Valeurs'")
```

```python
# DataFrame avec plusieurs colonnes
df = pd.DataFrame({
    'A': np.random.normal(0, 1, 1000),
    'B': np.random.normal(2, 1.5, 1000),
    'C': np.random.normal(-1, 2, 1000)
})

# Histogramme pour toutes les colonnes
df.plot(kind='hist', bins=30, alpha=0.5, title="Histogrammes de A, B et C")
```

```python
import pandas as pd
import numpy as np

# Exemple de DataFrame
df = pd.DataFrame({
    'Maths': np.random.normal(70, 10, 100),
    'Physique': np.random.normal(65, 15, 100),
    'Chimie': np.random.normal(75, 5, 100)
})

# Boxplot pour les colonnes numériques
df.plot(kind='box', title="Boxplot des notes")
```
