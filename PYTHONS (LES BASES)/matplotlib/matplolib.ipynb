# les 5 graphes les plus utilises en ML

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import  load_iris
```

(5) graphe de classification avec plt.scatter


```python
# telechargement de la donnee les fleurs d'load_iris

iris = load_iris()
x = iris.data
y = iris.target

# utilisation des donnees 
    # on rentre juste les variables et scatter se charge de tracer nos points.
    # alpha permet de regler la transparence des points
    # c = y color les points en fonction des targets(classes)
    # s = x[:,2]*10 permet de gerer la taille des points en fonction du toroisieme parametre

plt.scatter(x[:, 0], x[:,1], alpha= 0.5, c=y, s=x[:,2]*10)
plt.xlabel("longueur de sepal")
plt.ylabel("largeur des sepal")
```

```python

from mpl_toolkits.mplot3d import Axes3D
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(x[:,0],x[:,1],x[:,2], c=y)
```

```python
###  on peut utiliser la 3d pour visualiser une function'

# notre function
f = lambda x, y : np.sin(x) + np.cos(x + y)* np.cos(x)

# on commence d'abord par creer deux axes
X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)

# on cree notre grille en function de X Y
X, Y = np.meshgrid(X, Y)


Z = f(X,Y)
Z.shape
```

```python
### cmap = "plasma" pour la couleur
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, cmap = "plasma")
"""ax.set(xticklabels=[],
      yticklabels =[],
      zticklabels =[])"""
```

(3) Histogrammme

```python
plt.hist(x[:, 0], bins=20)
# bins un parametre qui permet de regler le nombre de section afficher 
# par defaut bins est a 10
```

```python
plt.hist(x[:, 0], bins=15)
plt.hist(x[:, 1], bins=15)
```

```python
plt.hist2d(x[:,0], x[:,1], cmap="Blues")
plt.colorbar() # indique a qoui servent les differentes couleurs 
plt.xlabel("Longueur de sepal")
plt.ylabel("largueur de sepal")
```

```python
from scipy import misc

face = misc.face(gray=True)
plt.hist(face.ravel(), bins=255)
```

(2) les contour plot

```python
# nb: x,y,z provienne de notre fonction d'en haut
plt.contour(X,Y,Z, 20, colors='Black') # 20 egale au nombre de niveau. plus la niveau est elevee plus 
# le graphe est detaille et visible.
```

```python
plt.contourf(X,Y,Z,20,cmap="gray")
plt.colorbar()
```
