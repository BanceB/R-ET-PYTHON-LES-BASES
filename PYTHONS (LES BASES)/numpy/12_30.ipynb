# NAN CORRECTIONS

```python
A = np.random.randn(5,5)
A[0, 2] = np.nan
A[4, 3] = np.nan

A
```

toutes les methodes statistiques vues ne fonctionnent que si 
nos donnees ne sont pas corrompues, c'est a dire qu'il n'y a pas de donnees manquantes ou que l'expression NAN N'est pas dans nous donees.
pour resoudre cela, on peut :
    1- utiliser des functions de calcule qui permet d'ignore ces valeurs manquantes (CES FONCTIONS COMMENCE PAR nan)
    2- remplacer ces valeurs innaproprye en valeur en passant par la methode isnan()

# 1

```python
# 1- 

# la moyenne
np.nanmean(A)
```

```python
# la varience

np.nanvar(A)
```

```python
# l'ecart_type

np.nanstd(A)
```

# 2

```python
 # 2-
 
np.isnan(A) # return True pour tous la nan
```

#### isnan() peut etre associer a d'autre fonctions

```python
np.isnan(A).sum() # determine le nombre de nan disponible
```

```python
np.isnan(A).sum() / A.size # determiner leurs pourcentages dans nos donnees
```

#### correction des nan

```python
# on remplace tous le nan par 0
A[np.isnan(A)] = 0
A
```

D'une maniere generale quand on veux faire de l'algebre lineaire on a recours a un subpackage totalement dedier a ca (np.linalg)

```python
A = np.random.randint(0,10,[3,3])
A
```

```python
# calcule du determinat 

np.linalg.det(A)
```

```python
# calcule de l'inverse de la matrice

np.linalg.inv(A)
```

```python
# determiner les valeurs propres et les vecteurs propres
        # souvent utiliser dans certains algorithmes(principal component analysis) pour faire de la 
        # reduction de dimension 
np.linalg.eig(A)
```
